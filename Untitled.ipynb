```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.image import imread
```

```python
image_raw = imread("cat.jpg")
print(image_raw.shape)
image_sum = image_raw.sum(axis=2)
print(image_sum.shape)
image_bw = image_sum/image_sum.max()
print(image_bw.max())
```

Original and B&W images 

```python
plt.subplot(1, 2, 1)
plt.imshow(image_raw)
plt.title('Original Image')

# B&W
plt.subplot(1, 2, 2)
plt.imshow(image_bw, cmap='gray')
plt.title('Black & White Image')

plt.show()
```

standardization

```python
mean = np.mean(image_bw)
std = np.std(image_bw)
imagestd = (image_bw - mean)/std
```

Covariance matrix 

```python
cov_matrix = np.cov(imagestd, rowvar=False)
cov_matrix
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
eigenvalues_sorted = sorted(eigenvalues, reverse=True)
eigenvectors_sorted = np.sort(eigenvectors)[::-1]
```

```python
print(f' Sorted eigen values:\n\n {eigenvalues_sorted} \n\n Sorted eigen vectors:\n\n{eigenvectors_sorted}')
```

```python
var_explained = [eigen_value/sum(eigenvalues) for eigen_value in eigenvalues_sorted]
cumulative_variance = np.cumsum(var_explained) *100
k = np.argmax(cumulative_variance>95)

'''Plot cumulative variance'''
plt.subplots(figsize=(10,5))
plt.grid()
for i in range(len(var_explained)):
    plt.axvline(x=k, color="k", linestyle="--")
    plt.axhline(y=95, color="r", linestyle="--")
plt.title('Cumulative variance') 
plt.xlabel('Number of principal components')
plt.ylabel('Cumulative variance')
ax = plt.plot(cumulative_variance)
plt.show()
```

```python
first3_eigenvectors = eigenvectors[:,:3]
principal_components = np.dot(image_bw,first3_eigenvectors)
principal_components
```

```python
from sklearn.decomposition import PCA, IncrementalPCA
ipca = IncrementalPCA(n_components=k)
image_recon = ipca.inverse_transform(ipca.fit_transform(image_bw))

plt.figure(figsize=[12,8])
plt.imshow(image_recon,cmap = plt.cm.gray)
```

```python
def plot_at_k(k):
    ipca = IncrementalPCA(n_components=k)
    image_recon = ipca.inverse_transform(ipca.fit_transform(image_bw))
    plt.imshow(image_recon,cmap = plt.cm.gray)

ks = [5, 15, 25, 75, 100, 170]

plt.figure(figsize=[15,9])

for i in range(6):
    plt.subplot(2,3,i+1)
    plot_at_k(ks[i])
    plt.title("Components: "+str(ks[i]))

plt.subplots_adjust(wspace=0.2, hspace=0.0)
plt.show()
```
